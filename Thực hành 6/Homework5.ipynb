```python
def Amstrong(a)->bool:
    do_dai=len(str(a))
    temp=a
    total=0
    while temp>0:
        chu_so=temp%10
        total+=chu_so**do_dai
        temp//=10
    return total==a
x=int(input())
print(Amstrong(x))
```

```python
import math
def quadratic_roots_count(a,b,c):
    delta=b**2-4*a*c
    if delta>0:
        n1=(-b-math.sqrt(delta))/(2*a)
        n2=(-b+math.sqrt(delta))/(2*a)
        return 'Hai nghiệm phân biệt',int(n1),int(n2)
    elif delta<0:
        return 'Vô Nghiệm'
    else:
        n0=-b/(2*a)
        return 'Nghiệm kép',int(n0) 
m,n,p=map(int,input().split())
print(quadratic_roots_count(m,n,p))
```

# W5A25

```python
import math
def triangle_area(x1, y1, x2, y2, x3, y3):
    return abs(x1*(y2 - y3) + x2*(y3 - y1) + x3*(y1 - y2)) / 2.0
def point_in_quadrilateral(quad, p):
    px, py = p
    (ax, ay), (bx, by), (cx, cy), (dx, dy) = quad
    area_quad = triangle_area(ax, ay, bx, by, cx, cy) + triangle_area(ax, ay, cx, cy, dx, dy)
    area1 = triangle_area(px, py, ax, ay, bx, by)
    area2 = triangle_area(px, py, bx, by, cx, cy)
    area3 = triangle_area(px, py, cx, cy, dx, dy)
    area4 = triangle_area(px, py, dx, dy, ax, ay)
    sum_areas = area1 + area2 + area3 + area4
    return math.isclose(area_quad, sum_areas)
quad = [(0, 0), (4, 0), (4, 3), (0, 3)]
p = (2, 1)
print(point_in_quadrilateral(quad, p))
```

```python
lst=list(map(int,input().split()))
def check(n):
    m=abs(n)
    total=0
    while m>0:
        chu_so=m%10
        if chu_so!=0:
            total+=1
        m//=10
    return total
def slcs(a):
    cnt=0
    for i in range(len(a)):
        if check(a[i])!=0:
            cnt+=check(a[i])
    return cnt
print(slcs(lst))
```

# W5A28

```python
import math
def goc_ba_diem(x1,y1,x2,y2,x3,y3):
    BAx=x1-x2
    BAy=y1-y2
    BCx=x3-x2
    BCy=y3-y2
    lenBA=math.sqrt(BAx**2+BAy**2)
    lenBC=math.sqrt(BCx**2+BCy**2)
    tvh=BAx*BCx+BAy*BCy
    cos_B=tvh/(lenBA*lenBC)
    goc_rad=math.acos(cos_B)
    goc_do=math.degrees(goc_rad)
    return goc_do
print(goc_ba_diem(0,1,0,0,1,0))
```

# W5A30

```python
def giaithua(a):
    total=1
    for i in range(2,a+1):
        total*=i
    return total
def hsnt(n,k):
    hs=giaithua(n)/(giaithua(k)*giaithua(n-k))
    return hs
n,k=map(int,input().split())
print(hsnt(n,k))
```

```python
def sum_1_to_n(n):
    if n<=0:
        return 0
    total=0
    for x in range(1,n+1):
        lst=[]
        check= True
        for digit in str(x):
            if digit in lst:
                check = False
                break
            lst.append(digit)
        if check:
            total+=x
    return total
n=int(input())
print(sum_1_to_n(n))
```
