```python
def Amstrong(a)->bool:
    do_dai=len(str(a))
    temp=a
    total=0
    while temp>0:
        chu_so=temp%10
        total+=chu_so**do_dai
        temp//=10
    return total==a
x=int(input())
print(Amstrong(x))
```

```python
def doixung(a)->bool:
    temp=a
    daonguoc=0
    while temp>0:
        chu_so=temp%10
        daonguoc=daonguoc*10+chu_so
        temp//=10
    return daonguoc==a
x=int(input())
print(doixung(x))
```
