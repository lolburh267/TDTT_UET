# W5A3

```python
import math
def nt(a):
    if n<2:
        return False
    else:
        for i in range(2,int(math.sqrt(a))+1):
            if a%i==0:
                return False
        return True
n=int(input())
if nt(n):
    print("Là số nguyên tố")
else:
    print("Không là số nguyên tố")
```

# W5A4

```python
import math
def shh(a)->int:
    tong=1
    for i in range(2,int(math.sqrt(a))+1):
        if a%i==0:
            tong+=i
            k=a//i
            if k!=i:
                tong+=k
    return tong
n=int(input())
if shh(n)==n:
    print("Là số hoàn hảo")
else:
    print("Không là số hoàn hảo")
```

# W5A5

```python
def timvitri(ds,k):
    for i in range(len(ds)):
        if ds[i]==k:
            return i
    return -1
nhap=input()
ds=[int(x) for x in nhap.split()]
k=int(input())
print(timvitri(ds,k))
```

# W5A7

```python
def maytinh(a,b,c):
    kq=0
    if c == "-":
        kq=a-b
    elif c == "+":
        kq=a+b
    elif c =="/":
        if b==0:
            return "Lỗi không thể chia cho 0"
        kq=a/b
    elif c=="*":
        kq=a*b
    return round(kq,2)
m,n=map(float,input().split())
num=input("Nhập toán tử cần thực hiện")
print(maytinh(m,n,num))
```

# W5A9

```python
def tongchuso(a):
    cnt=0
    while a>0:
        check=a%10
        cnt+=check
        a//=10
    return cnt
m=int(input())
print(tongchuso(m))
```

# W5A10

```python
def dangcau(a,b):
    if len(a)!=len(b):
        return False
    map_a_to_b={}
    map_b_to_a={}
    for i in range(len(a)):
        char_a=a[i]
        char_b=b[i]
        if char_a in map_a_to_b:
            if map_a_to_b[char_a]!=char_b:
                return False
        else:
            map_a_to_b[char_a]=char_b
        if char_b in map_b_to_a:
            if map_b_to_a[char_b]!=char_a:
                return False
        else:
            map_b_to_a[char_b]=char_a
    return True
m,n=map(str,input().split())
print(dangcau(m,n))
```

```python
def is_right_triangle(a,b,c):
    if a**2+b**2==c**2 or b**2+c**2==a**2 or a**2+c**2==b**2:
        return True
    else:
        return False
m,n,p=map(int,input().split())
print(is_right_triangle(m,n,p))
```
