# W6A16

```python
def main():
    n,m=map(int,input().split())
    matrix=[]
    for i in range(n):
        row=list(map(int,input().split()))
        matrix.append(row)
    for i in matrix:
        for j in i:
            print(f"{j:>4}",end="")
        print()
if __name__=='__main__':
    main()
```

# W6A20

```python
def sct(lst,k):
    for i in range(len(lst)):
        if lst[i]==k:
            return i
        else:
            return -1
lst=list(map(int,input().split()))
k=int(input())
print(sct(lst,k))
```

# W6A21

```python
n=input().split()
kq={}
for i in range(len(n)):
    if n[i] not in kq:
        kq[n[i]]=[i]
    else:
        kq[n[i]].append(i)
for key in kq:
    kq[key]=tuple(kq[key])
print(kq)
```

# W6A23

```python
def xcs(lst,x):
    kq=[]
    for i in range(len(lst)):
        if lst[i]>=x:
            kq.append(lst[i])
    return kq
lst=list(map(int,input().split()))
x=int(input())
print(xcs(lst,x))
```

```python
def dsk(lst):
    so_khong=[]
    check=[]
    kq=[]
    for i in lst:
        if i==0:
            so_khong.append(i)
        else:
            check.append(i)
    kq=so_khong+check
    return kq
lst=list(map(int,input().split()))
print(dsk(lst))
```
