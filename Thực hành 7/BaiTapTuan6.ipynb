# W6A9

```python
def input_dict(a):
    d={}
    nhap=input(a).split(',')
    for i in nhap:
        key,value=i.split(':')
        d[key.strip()]=[int(value.strip())]
    return d
dict1=input_dict()
dict2=input_dict()
check=dict1.copy()
for key,value in dict2.nhap():
    if key in check:
        check[key]+=value
    else:
        check[key]=value
sapxep=sorted(check.keys())
final={key:check[key]for key in sapxep}
print(final)
```

# W6A10

```python
lst=list(map(int,input().split()))
k=int(input())
kq=[]
n=len(lst)
for i in range(n):
    for j in range(i+1,n):
        if lst[i]+lst[j]==k:
            kq.append((lst[i],lst[j]))
m=len(kq)
for i in range(m):
    for j in range(0,m-i-1):
        if kq[j][0]>kq[j+1][0]:
            kq[j],kq[j+1]=kq[j+1],kq[j]
print(kq)
```

# W6A12

```python
lst=list(map(int,input().split()))
cnt={}
for i in range(len(lst)):
    if lst[i] in cnt:
        cnt[lst[i]]+=1
    else:
        cnt[lst[i]]=1
gtln=0
for i in cnt.values():
    if i > gtln:
        gtln=i
check=[]
for i,j in cnt.items():
    if j==gtln:
        check.append(i)
kq=check[0]
for i in check:
    if i<kq:
        kq=i
print(kq)
```

# W6A14

```python
data1=list(map(int,input().split()))
data2=list(map(int,input().split()))
kq=[]
for x in data1:
    check1=False
    for y in data2:
        if x==y:
            check1=True
            break
    check2=False
    for z in kq:
        if x==z:
            check2=True
            break
    if check1==True and check2==False:
        kq.append(x)
n=len(kq)
for i in range(n):
    for j in range(0,n-i-1):
        if kq[j]>kq[j+1]:
            kq[j],kq[j+1]=kq[j+1],kq[j]
print(kq)
```
