```python
a=input().split()
numbers=[int(x) for x in a]
result=[]
for num in numbers:
    if num not in result:
        result.append(num)
print(result)
```

```python
lst=list(map(int,input().split()))
kq=[]
total = 0
for i in range(len(lst)):
    total+=lst[i]
    kq.append(total)
print(kq)
```
